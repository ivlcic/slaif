# Inference and sampling

(Generative) language models estimate the probability of the next word given preceding context. To use these capable models for our task, we have to **sample from the output probability distribution**. As with most things in deep learning, the "correct" strategy is a matter of experimentation and validation. Below we demonstrate a few possible strategies.

An extended list of sampling hyperparameters supported in transformers is available at:  
https://huggingface.co/docs/transformers/v5.11.0/en/main_classes/text_generation#transformers.GenerationConfig

```python
import torch
from transformers import pipeline

# Use Text Generation pipeline with a decoder-only model (Qwen2.5 fine-tuned for chat)
pipeline = pipeline(task="text-generation",
                    model="Qwen/Qwen2.5-3B-Instruct",
                    dtype=torch.float16, device_map="auto")
```

Test use case: customer support assistant where the user input is commonly underspecified.

```python
chat = [
    {"role": "system", "content": "You are a customer support assistant responding with short but concrete answers."},
    {"role": "user", "content": "My internet keeps disconnecting."}
]
```

# Greedy decoding
Select the most probable word as the next one (set `do_sample=False`). Usually results in "overly safe" responses with no room for exploration.

```python

for _ in range(3):
    response = pipeline(chat, max_new_tokens=64, do_sample=False)
    print(response[0]["generated_text"][-1]["content"])
```

# Top-p sampling
Sample from most likely words whose cumulative probability minimally exceeds `p`. Enables reasonable exploration while cutting off very unlikely candidates.

```python
for _ in range(3):
    response = pipeline(chat, max_new_tokens=64, do_sample=True, top_p=0.95)
    print(response[0]["generated_text"][-1]["content"])
```

# Top-k sampling
Sample from k most probable words. Similar to top-p sampling, but sets limit based on number of most probable words instead of their cumulative probability.

```python
for _ in range(3):
    response = pipeline(chat, max_new_tokens=64, do_sample=True, top_k=5)
    print(response[0]["generated_text"][-1]["content"])
```
